### Visualization of Human Gait Force Data

Human Gait Analysys with Machine Learning is a good topic nowadays.I write this notebook to visualize the data and see "if it is a well-data" or "do we need pre-processing?". The results of graphs says: We do not need to pre-process data beacuse it's already done :)

```python
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def datanormalizer(data):
    a = 1
    b = 0
    c = 0
    newData = pd.DataFrame(index=np.arange(0,60),columns=['FP1_x','FP2_x','FP1_y','FP2_y','FP1_z','FP2_z'])
    for each in data.index:
        a = a+1
        if a%1000 == 0:
            newData.iloc[c] = data.iloc[b:a,:].sum()/1000
            b = a
            c = c+1
    return newData
data = pd.read_csv('GP1_0.6_force.csv')
data.head()
```

Lets visualize the force data in x direction

```python
timer = np.arange(0,60000)
plt.scatter(timer,data['FP1_x'],color="red")
plt.scatter(timer,data['FP2_x'],color="blue")
plt.xlabel('mS(time)')
plt.ylabel('Force')
plt.title('x-axis force change relative to time')
plt.grid()
plt.show()
```

As you see above, the right and left foots are work reversible.Lets look y-axis

```python
plt.scatter(timer,data['FP1_y'],color="red")
plt.scatter(timer,data['FP2_y'],color="blue")
plt.xlabel('mS')
plt.ylabel('Force')
plt.title('y-axis force change due to time')
plt.grid()
plt.show()
```

It's so normal that the right and left foot looks reflection of each other.So, let's check z axis too.

```python
plt.scatter(timer,data['FP1_z'],color="red")
plt.scatter(timer,data['FP2_z'],color="blue")
plt.xlabel('mS')
plt.ylabel('Force')
plt.title('z-axis force change due to time')
plt.grid()
plt.show()
```

We visualize the x,y and z axis of force.Also we can visualize the axis in a graph together too :)

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(data['FP1_x'],data['FP1_y'], data['FP1_z'], color='red', marker='o')
ax.scatter(data['FP2_x'],data['FP2_y'], data['FP2_z'], color='blue', marker='.')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title("3D Left-Right Foot Point Cloud")
plt.show()
```

I also write a function to visualize data with lower samples to see more basicly.

```python
q = datanormalizer(data=data)

plt.scatter(np.arange(0,60),q['FP1_x'])
plt.scatter(np.arange(0,60),q['FP2_x'])
plt.title('reduced data')
plt.xlabel('S')
plt.ylabel('x_basinc')
plt.show()
```

### Visualization of Human Foot Position in 3D

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('GP1_0.6_marker.csv')
data.head()
```

```python
import pandas as pd
import numpy as np
```

Let's import the data of 1st subject with 0.6 and 0.7 m/s speed.

```python
datasix1 = pd.read_csv('GP1_0.6_marker.csv')
dataseven2 = pd.read_csv('GP1_0.7_marker.csv')
datasix1.head()
```

Combine both 0.7 and 0.6 speed data with class of 0.6 m/s speed data is 0 and class of 0.7 m/s speed data is 1. dependent = class

```python
data1 = pd.DataFrame({'dependent':np.zeros(12000)})
data2 = pd.DataFrame({'dependent':np.ones(12000)})

data2 = pd.concat([dataseven2,data2],axis=1)
data1 = pd.concat([datasix1,data1],axis=1)

data = data1.append(data2,ignore_index=True)
data.head()
```

Y = dependent = classes , x = data without classes

```python
x = data.drop(["dependent"],axis=1)
y = data.dependent.values
x.head()
```

Let's split the data as train and test.

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

#### Feature Scaling




```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(x_train)
X_test_sc = sc.transform(x_test)
```

### Logistic Regression


```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_train_sc,y_train)

predictions = logmodel.predict(X_test_sc)

from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, predictions)
print(cm)
accuracy_score(y_test, predictions)
```

```python
#K-Cross Validation #Updated Logistic Regression with best Hyperparameters Selected

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = logmodel, X = X_train_sc, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier1 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier1.fit(X_train_sc, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred1 = classifier1.predict(X_test_sc)
cm = confusion_matrix(y_test, y_pred1)
print(cm)
accuracy_score(y_test, y_pred1)
```

```python

import pandas as pd
import numpy as np

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```
